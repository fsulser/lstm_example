```python
stop_words = set(['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', 'her', 'hers', 'herself', 'it', 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', 'these', 'those', 'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through', 'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further', 'then', 'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other', 'some', 'such', 'no', 'nor', 'not', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 's', 't', 'can', 'will', 'just', 'don', 'should', 'now']) 


def clean_string(sent): 
  words = str(sent).lower().split(' ')

  #Removing all single letter and and stopwords from question 
  #use stemming as well
#  sent1=' '.join(str(stemmer.stem(j)) for j in words if j not in stop_words and (len(j)!=1)) 

  return ' '.join(str(j) for j in words if j not in stop_words and (len(j)!=1)).replace('::', ' ')
```

```python
#clean inputs
df['text1_cleaned'] = df.apply(lambda x: clean_string(x['text1']), axis=1)
df['text2_cleaned'] = df.apply(lambda x: clean_string(x['text2']), axis=1)
```

```python
max_len = 64

all_text = df[['text1_cleaned']].apply(lambda x:str(x[0]) + " ", axis=1).tolist()
all_text.extend(df[['text2_cleaned']].apply(lambda x:str(x[0]) + " ", axis=1).tolist())

t = Tokenizer()
t.fit_on_texts(all_text)
```

```python
#tokenize strings
df['text1_seq'] = t.texts_to_sequences(df['text1_cleaned'].astype(str).values)
df['text2_seq'] = t.texts_to_sequences(df['text2_cleaned'].astype(str).values)
```

```python
#sequence padding
df['text1_seq_pad'] = [list(x) for x in pad_sequences(df['text1_seq'], maxlen=max_len, padding='post')]
df['text2_seq_pad'] = [list(x) for x in pad_sequences(df['text2_seq'], maxlen=max_len, padding='post')]
```

```python
def l1_distance(x):
  return backend.abs(x[0] - x[1])

def l1_dist_output_shape(x):
  return x[0]
  
    
common_embed = Embedding(
    name="synopsis_embedd",
    input_dim=len(t.word_index)+1,
    output_dim=300,
    input_length=len(df['text1_seq_pad'].tolist()[0]),
    trainable=True
)

lstm_layer = tf.keras.layers.Bidirectional(
    tf.keras.layers.LSTM(32, dropout=0.2, recurrent_dropout=0.2)
)

input1 = tf.keras.Input(shape=(len(df['text1_seq_pad'].tolist()[0]),))
e1 = common_embed(input1)
x1 = lstm_layer(e1)

input2 = tf.keras.Input(shape=(len(df['text2_seq_pad'].tolist()[0]),))
e2 = common_embed(input2)
x2 = lstm_layer(e2)

merged = tf.keras.layers.Lambda(
    function=l1_distance, output_shape=l1_dist_output_shape, name='L1_distance'
)([x1, x2])

conc = Concatenate(axis=-1)([merged, x1, x2])

x = Dropout(0.01)(conc)
preds = tf.keras.layers.Dense(1, activation='sigmoid')(x)
model = tf.keras.Model(inputs=[input1, input2], outputs=preds)
```

```python

METRICS = [
  tf.keras.metrics.TruePositives(name='tp'),
  tf.keras.metrics.FalsePositives(name='fp'),
  tf.keras.metrics.TrueNegatives(name='tn'),
  tf.keras.metrics.FalseNegatives(name='fn'), 
  tf.keras.metrics.BinaryAccuracy(name='accuracy'),
  tf.keras.metrics.Precision(name='precision'),
  tf.keras.metrics.Recall(name='recall'),
  tf.keras.metrics.AUC(name='auc'),
  tf.keras.metrics.AUC(name='prc', curve='PR'),
]
```

```python
epochs = 30
batch_size = 4


loss = tf.keras.losses.BinaryCrossentropy(from_logits=True)
model.compile(loss=loss, metrics=METRICS, optimizer='adam')


model.fit(
    x = [np.array(df['text1_seq_pad'].tolist()), np.array(df['text2_seq_pad'].tolist())],
    y = df['belongs_together'].values.reshape(-1,1), 
    epochs=epochs,
    batch_size=batch_size,
    validation_data=([np.array(df['text1_seq_pad'].tolist()), np.array(df['text2_seq_pad'].tolist())], df['belongs_together'].values.reshape(-1,1))
)
```
